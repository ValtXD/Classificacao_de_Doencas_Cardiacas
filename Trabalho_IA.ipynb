```python
# Importa a função train_test_split para dividir os dados em treino e teste
from sklearn.model_selection import train_test_split

# Separa os atributos (X) e a variável alvo (y)
X = df.drop('target', axis=1)   # Remove a coluna 'target' e usa o restante como entrada
y = df['target']                # Define a coluna 'target' como o valor a ser previsto

# Divide os dados em treino e teste (80% treino, 20% teste)
# 'stratify=y' mantém a proporção de classes (0 e 1) igual nos dois conjuntos
# 'random_state=42' garante reprodutibilidade dos resultados
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# Exibe o tamanho do conjunto de treino
print("Tamanho do conjunto de treino:", X_train.shape)

# Exibe o tamanho do conjunto de teste
print("Tamanho do conjunto de teste:", X_test.shape)
```

Nesta etapa fundamental do processo de modelagem preditiva, procedeu-se à divisão estruturada do dataset em conjuntos de treinamento e teste. Inicialmente, importou-se a função train_test_split da biblioteca scikit-learn, seguida pela separação lógica entre variáveis preditoras (X) e variável target (y). A coluna 'target' foi removida do conjunto de features e designada como variável resposta, mantendo as 13 restantes como preditores. Configurou-se a divisão com proporção 80-20, utilizando o parâmetro stratify=y para preservar a distribuição balanceada das classes e random_state=42 para garantir completa reprodutibilidade dos resultados.

Visualizações Geradas:

Tamanho do conjunto de teste: (205, 13)

A divisão foi executada com sucesso, resultando em 820 amostras para treinamento (80% do total) e 205 para teste (20%), mantendo todas as 13 variáveis preditoras originais. A estratificação garantiu que a proporção balanceada de 50% para cada classe (presença e ausência de doença cardíaca) fosse preservada em ambos os conjuntos, assegurando que o modelo seja treinado e avaliado em distribuições representativas da população original. Esta abordagem preventiva contra viés de seleção é crucial para o desenvolvimento de modelos generalizáveis.

```python
# Importa o StandardScaler para padronizar (normalizar) os dados
from sklearn.preprocessing import StandardScaler

# Cria o objeto de padronização
scaler = StandardScaler()

# Ajusta o scaler aos dados de treino e transforma (normaliza) esses dados
X_train_scaled = scaler.fit_transform(X_train)

# Usa o mesmo scaler para transformar os dados de teste (sem ajustar novamente)
X_test_scaled = scaler.transform(X_test)

# Mostra a média dos dados normalizados de treino (deve ser próxima de 0)
print("Treino normalizado - média:", X_train_scaled.mean().round(2))

# Mostra o desvio padrão dos dados normalizados de treino (deve ser próximo de 1)
print("Treino normalizado - desvio padrão:", X_train_scaled.std().round(2))
```

Prosseguindo com o pré-processamento, implementou-se a padronização das variáveis através do StandardScaler do scikit-learn. Criou-se o objeto scaler e aplicou-se a transformação seguindo as melhores práticas de machine learning: o scaler foi ajustado e transformado exclusivamente nos dados de treinamento usando fit_transform(), enquanto os dados de teste foram transformados usando a mesma parametrização através de transform(), prevenindo vazamento de informação. Realizou-se verificação estatística pós-padronização para validar a eficácia do processo.

Visualizações Geradas:

A padronização foi executada com precisão, reduzindo a média para aproximadamente zero e o desvio padrão para 1.0, confirmando que todas as variáveis foram adequadamente centralizadas e escaladas. Esta transformação é particularmente crítica considerando a ampla variabilidade de escalas identificada na análise exploratória anterior - desde idade (20-80 anos) até colesterol (126-564 mg/dl) e oldpeak (0-6.2). A abordagem conservadora de ajustar o scaler apenas nos dados de treinamento simula fielmente condições reais de implantação, onde novos dados seriam processados com base na distribuição dos dados históricos, estabelecendo bases sólidas para modelos preditivos robustos.

```python
# Importa o TensorFlow e os módulos necessários para criar e treinar a rede neural
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam

# Criação do modelo sequencial (camadas empilhadas)
model = Sequential([
    # Primeira camada densa com 64 neurônios e ativação ReLU (entrada)
    Dense(64, activation='relu', input_shape=(X_train_scaled.shape[1],)),

    # Camada Dropout para evitar overfitting (remove 30% dos neurônios aleatoriamente durante o treino)
    Dropout(0.3),

    # Segunda camada oculta com 32 neurônios e ativação ReLU
    Dense(32, activation='relu'),

    # Outra camada Dropout (remove 20% dos neurônios)
    Dropout(0.2),

    # Terceira camada oculta com 16 neurônios e ativação ReLU
    Dense(16, activation='relu'),

    # Camada de saída com 1 neurônio e ativação sigmoide (para classificação binária)
    Dense(1, activation='sigmoid')
])

# Compilação do modelo — define o otimizador, a função de perda e as métricas de avaliação
model.compile(
    optimizer=Adam(learning_rate=0.001),  # Usa o otimizador Adam com taxa de aprendizado de 0.001
    loss='binary_crossentropy',           # Função de perda usada em classificações binárias
    metrics=['accuracy']                  # Métrica de desempenho: acurácia
)
```

### Arquitetura da Rede Neural

O modelo é uma rede neural *feedforward* com 13 neurônios de entrada (correspondentes aos atributos clínicos) e uma camada de saída com um único neurônio **sigmoide**, que retorna a probabilidade de o paciente ter doença cardíaca.

As camadas intermediárias utilizam a função de ativação **ReLU** e a técnica de **Dropout** para evitar *overfitting*.

**Resumo da Arquitetura:**
- 1ª camada: 64 neurônios, ReLU  
- 2ª camada: 32 neurônios, ReLU  
- 3ª camada: 16 neurônios, ReLU  
- Camada de saída: 1 neurônio, Sigmoid  
- Função de perda: Binary Crossentropy  
- Otimizador: Adam

Nesta etapa de construção do modelo de rede neural, procedeu-se à importação e implementação de uma arquitetura deep learning utilizando TensorFlow/Keras. A rede neural foi estruturada como um modelo sequencial composto por quatro camadas densas intercaladas com camadas de dropout para regularização. Configurou-se a primeira camada com 64 neurônios e função de ativação ReLU, definindo explicitamente o input_shape correspondente às 13 variáveis preditoras do dataset escalado. Adicionou-se uma camada de dropout de 30% seguida por uma segunda camada oculta com 32 neurônios ReLU, outra camada de dropout de 20%, uma terceira camada com 16 neurônios ReLU, e finalmente uma camada de saída com 1 neurônio e ativação sigmoide para classificação binária. O modelo foi compilado com otimizador Adam (learning_rate=0.001), função de perda binary_crossentropy e métrica de acurácia.

A saída do model.summary() apresenta uma tabela detalhada da arquitetura da rede neural:

A arquitetura construída apresenta uma estrutura completamente esperada e adequada para o problema de classificação binária em questão. A progressão decrescente de neurônios (64 → 32 → 16 → 1) segue um padrão piramidal convencional em redes neurais, permitindo que a rede aprenda features hierárquicas gradualmente. O número total de 3.521 parâmetros é considerado moderado para o tamanho do dataset (820 amostras de treino), indicando boa relação entre capacidade de modelagem e risco de overfitting. A inclusão estratégica de camadas dropout com taxas de 30% e 20% demonstra preocupação com regularização, essencial para evitar sobreajuste em problemas médicos com dados limitados. A configuração de 896 parâmetros na primeira camada calcula-se corretamente como (13 features × 64 neurônios) + (64 bias) = 896, validando a implementação. A camada de saída com ativação sigmoide e 1 neurônio é a escolha padrão e correta para classificação binária. Esta arquitetura balanceia complexidade e generalização, sendo plenamente adequada para a predição de doença cardíaca.

```python
# Importa o callback EarlyStopping para interromper o treino automaticamente
from tensorflow.keras.callbacks import EarlyStopping

# Cria o EarlyStopping para parar o treino se a perda de validação não melhorar por 10 épocas
# 'restore_best_weights=True' faz o modelo voltar aos melhores pesos obtidos
es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Treina o modelo com os dados normalizados
history = model.fit(
    X_train_scaled, y_train,   # Dados de treino
    validation_split=0.2,      # Usa 20% dos dados de treino para validação
    epochs=100,                # Número máximo de épocas de treino
    batch_size=16,             # Quantidade de amostras por iteração
    callbacks=[es],            # Aplica o EarlyStopping durante o treino
    verbose=1                  # Exibe o progresso do treinamento no console
)
```

Nesta etapa crucial de treinamento da rede neural, implementou-se o processo de aprendizagem supervisionada com mecanismos avançados de controle. Configurou-se o callback EarlyStopping para monitorar continuamente a perda de validação (val_loss) com paciência de 10 épocas, permitindo interromper o treinamento precocemente caso não houvesse melhoria, enquanto restore_best_weights=True garantiu a manutenção dos melhores parâmetros encontrados. O modelo foi treinado por até 100 épocas utilizando os dados normalizados, com divisão de 20% do conjunto de treino para validação, batch_size de 16 amostras por iteração e acompanhamento detalhado do progresso através do parâmetro verbose=1.

O treinamento executou todas as 100 épocas configuradas, apresentando uma evolução notável nas métricas de desempenho. A acurácia de treino iniciou em 50.71% na primeira época, evoluindo consistentemente até estabilizar acima de 97% nas épocas finais. A perda de treino (loss) reduziu drasticamente de 0.7815 para aproximadamente 0.0641. Mais impressionante foi o desempenho na validação: a val_accuracy atingiu 100% consistentemente a partir da época 78, enquanto a val_loss convergiu para valores próximos de 0.0246, demonstrando excelente generalização. O EarlyStopping não foi acionado devido à contínua melhoria nas métricas.

Os resultados obtidos são excepcionais e, surpreendentemente, superam as expectativas iniciais. A convergência para 100% de acurácia na validação é um desempenho raro em problemas médicos reais e pode indicar tanto a alta qualidade do modelo quanto possíveis particularidades no conjunto de dados. A trajetória de aprendizagem mostra um padrão saudável: melhoria gradual e consistente tanto no treino quanto na validação, sem sinais evidentes de overfitting - fato atribuído à eficácia das camadas de Dropout implementadas. A estabilização das métricas nas épocas finais, com val_loss mantendo-se em patamares baixos (≈0.024) e val_accuracy em 100%, sugere que o modelo alcançou plena convergência.
```python
# Importa funções para calcular métricas de desempenho do modelo
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Gera previsões do modelo (valores entre 0 e 1)
y_pred_prob = model.predict(X_test_scaled)

# Converte as probabilidades em classes binárias (0 ou 1)
y_pred = (y_pred_prob >= 0.5).astype(int)

# Calcula a acurácia (proporção de acertos)
acc = accuracy_score(y_test, y_pred)

# Calcula a precisão (taxa de acertos entre as previsões positivas)
prec = precision_score(y_test, y_pred)

# Calcula o recall (taxa de acertos entre os casos positivos reais)
rec = recall_score(y_test, y_pred)

# Exibe as métricas principais formatadas com 4 casas decimais
print(f"Acurácia: {acc:.4f}")
print(f"Precisão: {prec:.4f}")
print(f"Recall: {rec:.4f}")

# Exibe um relatório completo com várias métricas (precisão, recall, F1-score e suporte)
print("\nRelatório de Classificação:")
print(classification_report(y_test, y_pred))
```
